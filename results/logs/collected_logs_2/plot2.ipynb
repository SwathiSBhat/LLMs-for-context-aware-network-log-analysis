```python
audit = open('./audit.log.new', 'r').readlines()
auth = open('./auth.log.new', 'r').readlines()
syslog = open('./syslog.new', 'r').readlines()
kern = open('./kern.log.new', 'r').readlines()



print('audit:', len(audit))
print('auth:', len(auth))
print('syslog:', len(syslog))
print('kern:', len(kern))
```

```python
data = [len(audit), len(auth), len(syslog), len(kern)]

plt.bar(['audit', 'auth', 'syslog', 'kern'], data)
plt.xlabel('Log Types')
plt.ylabel('Number of Entries')
plt.savefig('./log_types_2.pdf')

plt.show()
```
